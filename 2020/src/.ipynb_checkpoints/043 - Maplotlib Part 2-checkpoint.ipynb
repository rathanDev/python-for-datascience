```python
fig = plt.figure()
fig.savefig('../resources/my_pic.png', dpi=200)
ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y)
# ax.plot(y,x)
ax.plot(x,x**2,'r',label='X Squared')
ax.plot(x,x**3,'g', label='X Cubed')

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')

ax.legend(loc=10)
```
