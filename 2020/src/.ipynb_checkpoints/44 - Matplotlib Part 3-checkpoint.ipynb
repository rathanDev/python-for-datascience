```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='purple',lw=3,alpha=0.5) linestyle='steps',
ax.plot(x,y,color='purple',lw=3, marker='o', markeredgewidth=8)
```
